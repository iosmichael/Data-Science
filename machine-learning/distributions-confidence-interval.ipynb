Distribution is especially important when we use them to model variables in regressions, and being able to understand the implication behind different distributions can build a good foundation for further statistical studies.

**Gamma function**
$$\Gamma(n) = (n-1)!$$

```python
%matplotlib inline
from matplotlib import pyplot as plt
import scipy.stats as stats
import numpy as np
```

**Normal Distribution**

```python
x = np.arange(0, 20, 0.1)
y = stats.norm.pdf(x, 10, 3)
plt.plot(x, y, 'r')
plt.show()
```

**Chi-Square Distribution**

Normally used for categorical datasets. A distribution that represents square of a normal distributed random variable. 

The support of random variable is $\mathbf{O} > 0$.

$$\chi^2_c = \sum \frac{(O-E)^2}{E}$$

where $O$ is observed value and $E$ is expected value

**Normally, chi square distribution represents the square sum of $m$ independent random variables.** *(chi-square distribution with m degree of freedom)*

```python
x = np.arange(0, 20, 0.1)
y = stats.chi.pdf(x, 10, 3) #degree of freedom = 3
#compared with normal distribution
yn = stats.norm.pdf(x, 10, 3)
plt.plot(x, y, 'r', x, yn, 'b')
plt.show()
```

**T-Distribution**

A distribution that deal with unknown sample variance or small dataset ($N < 30$)

***Mathematical Proof:***

**Theorem 1** 

Let $U = \sum_{j=1}^m Z^2_j$, where $Z_1, Z_2, \ldots, Z_m$ are independent standard normal random variables. Then $U$ has a gamma distribution with $r=\frac{m}{2}$ and $\lambda = \frac{1}{2}$. That is,
$$f_{U}(u) = \frac{1}{2^{m/2}\Gamma(\frac{m}{2})}u^{(m/2)-1}e^{-u/2}, \ \ u\geq 0$$

*Since the distribution of the sum of squares of independent standard normal random variables is sufficiently important, we call this special case of Gamma distribution **"Chi Square Distribution"***

**Properties of this theorem:**

- $Z_1, Z_2, \ldots, Z_m$ are independent standard normal
- The pdf of U is called the *chi square distribution with **m** degrees of freedom*

**Theorem 2**
Let $Y_1, Y_2, \ldots, Y_n$ be a random sample from a normal distribution with mean $\mu$ and variance $\sigma^2$. Then,

- $S^2$ and $\bar{Y}$ are independent
- $\frac{(n-1)S^2}{\sigma^2} = \frac{1}{\sigma^2}\sum_{i=1}{n}(Y_i - \bar{Y})^2$ has a *chi square distribution with **n-1** degrees of freedom*.

**Theorem 3**
Suppose that U and V are independent chi square random variables with n and m degrees of freedom, respectively. A random variable of the form $\frac{V/m}{U/n}$ is said to have a **F distribution** with *m and n degrees of freedom*.


```python
df_l = [i for i in range(1,10,2)]
default = 10
x = np.linspace(stats.t.ppf(0.01, default), stats.t.ppf(0.99, default), 100)

fig, ax = plt.subplots(1,1)
#compared with standard z-distribution
ax.plot(x, stats.norm.pdf(x, 0, 1), 'b')
#plot t distribution with various degrees of freedom
for df in df_l:
    y = stats.t.pdf(x, df)
    ax.plot(x, y, 'r', alpha = 0.5, label="t "+str(df))
ax.set_title("T-Distribution vs Z-Distribution")
plt.show()
```

## Confidence Interval

### Single Variable

Using Z-distribution: 

The confidence interval of population mean $\mu$, using sample mean and population variance ($\hat{y}$, $\sigma^2$) with $\alpha$ level of significance. (two-sided confidence interval)

$$(\hat{y} + z_{\frac{\alpha}{2}}\frac{\sigma^2}{\sqrt{n}}, \hat{y} + z_{1-\frac{\alpha}{2}}\frac{\sigma^2}{\sqrt{n}})$$

**Interpretation of confidence interval (2, 4) with 95% significance:**

* *According to the evidence given in our observed statistical data, we can reason that there is 95% of probability that the population mean falls between 2 and 4.*

**Confidence Interval Assumptions**:
* Individual samples are identically and independently selected from the population sample space.
* The sample variance presumably equal to population variance $S^2 = \sigma ^2$
* The sample is large enough to assume central limited theorem. **CLT** (n > 30)

Using T-distribution:

The confidence interval of population mean $\mu$, using sample mean and sample variance ($\hat{y}$, $S^2$) with $\alpha$ level of significance. (two-sided confidence interval)

$$(\hat{y}+t_{\frac{\alpha}{2}}\frac{S^2}{\sqrt{n}}, \hat{y}+t_{1-\frac{\alpha}{2}}\frac{S^2}{\sqrt{n}})$$

**Confidence Interval Assumptions**:
* Individual samples are identically and independently selected from the population space.
* The sample variance has a chi-square relationship with the population variance
* The sample size is small and we do not assume central limit theorem. ( $n \leq 30$ )

Using Chi-square distribution:

**Inference about $\sigma^2$**

Chi-square distribution is specially useful when we draw inference about the population variance $\sigma^2$. The confidence interval of population variance $\sigma^2$, using sample variance ($S^2$) with $\alpha$ level of significance. (two-sided confidence interval)

**Confidence Interval on $\sigma^2$**

$$(\frac{(n-1)s^2}{\chi^2_{1-\frac{\alpha}{2},n-1}}, \frac{(n-1)s^2}{\chi^2_{\frac{\alpha}{2},n-1}})$$

**Confidence Interval $\sigma$**

$$(\frac{(n-1)s^2}{\chi^2_{1-\frac{\alpha}{2},n-1}}, \frac{(n-1)s^2}{\chi^2_{\frac{\alpha}{2},n-1}})$$

### Two Variables

**Inference about $X-Y$**

$$E(\mathbf{X} - \mathbf{Y}) = E(\mathbf{X})-E(\mathbf{Y})$$

**Pooled Sample Variance**

Assume that X and Y both have the **same population variance**, then: (This case proves to be very rare in the real world)

$$Var(X - Y) = \sigma^2/n + \sigma^2/m = S_p\sqrt{\frac{1}{n} + \frac{1}{m}}$$

where $S_p$ is pooled sample variance and $S_p^2 = \frac{\sum^{n}_{i=1}(X_i - \bar X)^2 + \sum^m_{i=1}(Y_i - \bar Y)^2}{n+m-2}$

*Its associated t-distribution has a degree of freedom of $n+m-2$.*

**Unpooled Sample Variance**

Assume that X and Y each has a distinctive population variance, then:

$$Var(X-Y) = \sqrt{Var(X)^2 + Var(Y)^2} = \sqrt{\frac{S_X^2}{n} + \frac{S_Y^2}{m}}$$

*Its associated t-distribution has a degree of freedom of $\frac{(\hat\theta + \frac{n}{m})^2}{\frac{1}{(n-1)}\hat\theta^2 + \frac{1}{m-1}(\frac{n}{m})^2}$.* where $\hat\theta = \frac{S_X^2}{S_Y^2}$

**Using F-distribution:**

F-distribution is used when we analyze the ratio between two chi-square random variables. Therefore, this is especially useful for drawing the inference of the ratio between two sample variables.

$$\frac{S_X^2}{S_Y^2} \propto \mathcal{F}_{n-1, m-1}$$
